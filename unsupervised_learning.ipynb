```python
#clustering techniques include KMeans,Hierachical clustering and Density based clustering(DBSCAN)

#importation 
from sklearn.cluster import KMeans,DBSCAN

#data splitting
x=iris[iris.columns[:-1]].values #samples generation--should be in a numpy array format


#KMeans
kmeans=KMeans(n_clusters=3) #n_clusters=number of clusters
kmeans.fit(x) #X is the dataset
predictions=kmeans.predict(x) #predicting the clusters

#print predictions
print(predictions)

#plot the predictions for each dimension in the sample
for i in range(1,5):

  plt.scatter(x[:,i],predictions,c=y)
  plt.title("Clustering using KMeans")
  plt.xlabel(i)
  plt.show()
```

```python
#To create a dendrogram using sklearn's AgglomerativeClustering, you'll need to combine it with scipy's dendrogram function. Here's a step-by-step guide on how to do this: [1]
from scipy.cluster.hierachy import dendrogram

import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram
from sklearn.cluster import AgglomerativeClustering

# Generate sample data
np.random.seed(0)
X = np.random.rand(20, 2)

def plot_dendrogram(model, **kwargs):
    # Create linkage matrix and then plot the dendrogram

    # create the counts of samples under each node
    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1  # leaf node
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count

    linkage_matrix = np.column_stack([model.children_, model.distances_,
                                      counts]).astype(float)

    # Plot the corresponding dendrogram
    dendrogram(linkage_matrix, **kwargs)

# Create the AgglomerativeClustering object
# Setting distance_threshold=0 ensures we compute the full tree
model = AgglomerativeClustering(distance_threshold=0, n_clusters=None)

# Fit the model
model = model.fit(X)

# Plot the dendrogram
plt.title('Hierarchical Clustering Dendrogram')
plot_dendrogram(model, truncate_mode='level', p=3)
plt.xlabel("Number of points in node (or index of point if no parenthesis).")
plt.show()
```
